```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

labels = pd.read_csv("drive/MyDrive/dog-breed-data/labels.csv")
print(labels.describe)
print(labels.head())
```

```python
labels.breed.value_counts()
```

```python
labels.breed.value_counts().plot(kind="bar", figsize=(20, 10))
```

```python
import os
len(os.listdir("/content/drive/MyDrive/dog-breed-data/train/")) == len(labels)
```

```python
rows = 3
cols = 4
row = 0
col = 1
fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(10, 8))
for i in range(12):
  id = labels.id[i]
  file = f"drive/MyDrive/dog-breed-data/train/{id}.jpg"
  image = plt.imread(file)
  if i == 0:
    ax[0, 0].imshow(image)
    ax[0, 0].set_title(labels.breed[i])
    ax[0, 0].axis("off")
  else:
    if i % cols == 0:
      col = 0
      row += 1
    ax[row, col].imshow(image)
    ax[row, col].set_title(labels.breed[i])
    ax[row, col].axis("off")
    col += 1
```
